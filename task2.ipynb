```python
import requests
import pandas as pd
import schedule
import time
from datetime import datetime

# ---- Configuration ----
API_KEY = 'f7cfe7d1ed0a7658c8f03ef7c1e552ffc1d422b4'  # Replace with your actual JCDecaux API key
CITY = 'lyon'  # Specify Lyon as the city
URL = f'https://api.jcdecaux.com/vls/v3/stations?contract={CITY}&apiKey={API_KEY}'
FILE_NAME = 'lyon_bike_data.csv'

# ---- Data Fetching Function ----
def fetch_data():
    try:
        response = requests.get(URL)
        response.raise_for_status()  # Raise exception for bad response
        data = response.json()

        # Convert JSON data to DataFrame
        df = pd.DataFrame(data)
        
        # Add timestamp to the data
        df['timestamp'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Save data to CSV (append mode)
        df.to_csv(FILE_NAME, mode='a', index=False, header=not pd.io.common.file_exists(FILE_NAME))
        print(f'Data for Lyon saved at {datetime.now()}')

    except Exception as e:
        print(f"Error fetching data: {e}")

# ---- Automate Periodic Fetching ----
schedule.every(20).minutes.do(fetch_data)  # Fetch every 20 minutes

print("Starting data collection for Lyon...")
while True:
    schedule.run_pending()
    time.sleep(1)
```
